```
!pip install librosa numpy pandas

import os
import librosa
import numpy as np
import pandas as pd


def preprocess_audio(audio_path,output_path):
  # Load an audio file
  audio_file = audio_path

  # Adjust the sampling rate to 100Hz (10ms intervals)
  y, sr = librosa.load(audio_file, sr=100)

  # Extract MFCCs
  mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=12)
  zcr = librosa.feature.zero_crossing_rate(y)
  spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
  spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)
  spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
  spectral_flatness = librosa.feature.spectral_flatness(y=y)
  chroma = librosa.feature.chroma_stft(y=y, sr=sr)
  harmonic, percussive = librosa.effects.hpss(y)
  hnr = librosa.effects.harmonic(y=harmonic)

  # Extract pitch
  f0, voiced_flag, voiced_probs = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))
  # print(f0)
  # Extract VUV (voiced/unvoiced) feature
  vuv = voiced_flag

  # Ensure all feature arrays have the same length

  min_length = min(len(mfccs[0]), len(f0), len(vuv),len(zcr[0]))

  # Create a list of lists for each frame
  data = []

  for i in range(min_length):
      frame_data = list(mfccs[:, i]) + [zcr[0][i],f0[i], vuv[i],spectral_centroid[0][i],spectral_bandwidth[0][i],spectral_rolloff[0][i],spectral_flatness[0][i]]
      # frame_data = list(chroma[i,:])
      data.append(frame_data)

  # Create a DataFrame
  df = pd.DataFrame(data, columns=[f'MFCC{i+1}' for i in range(12)] + ['ZCR','F0', 'VUV','spectral_centroid','spectral_bandwidth','spectral_rolloff','spectral_flatness'])

  # Save the DataFrame to a CSV file
  csv_file = output_path
  df.to_csv(csv_file, index=False)

  # print(df.shape)

  # print(f"Extracted features sampled at 100Hz saved to {csv_file}")

# Directories to process
directories = [
    "/content/drive/MyDrive/MCA Dataset/dev_data",
    "/content/drive/MyDrive/MCA Dataset/test_data",
    "/content/drive/MyDrive/MCA Dataset/train_data"
]

for dir_path in directories:
    # print(f"Processing directory: {dir_path}")

    # Check if directory exists
    if not os.path.exists(dir_path):
        print(f"Directory not found: {dir_path}")
        continue

    # List all files in the directory
    for file_name in os.listdir(dir_path):
        # Check if the current file is an audio file
        if file_name.endswith("_AUDIO.wav"):
            # Full path to the audio file
            audio_path = os.path.join(dir_path, file_name)
            print(f"Processing file: {audio_path}")

            # Output CSV file path (change extension to .csv)
            output_path = os.path.join(dir_path, file_name.replace(".wav", ".csv"))

            # Process the audio file and extract features
            try:
                preprocess_audio(audio_path, output_path)
            except Exception as e:
                print(f"An error occurred while processing {audio_path}: {e}")

    print(f"Completed processing directory: {dir_path}")

print("All directories have been processed.")
```
